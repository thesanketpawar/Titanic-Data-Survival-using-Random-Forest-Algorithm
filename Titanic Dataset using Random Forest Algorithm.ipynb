```python
import numpy as np
import pandas as pd
```

```python
data=pd.read_csv('/content/train.csv')
test=pd.read_csv('/content/test.csv')
```

```python
data.head()
```

```python
#Checking the Null Values
print(data.isnull().sum())
print(test.isnull().sum())
```

```python
data=data.drop(['PassengerId','Name','Ticket','Cabin'],axis=1)
test=test.drop(['PassengerId','Name','Ticket','Cabin'],axis=1)
```

```python
data.head()
```

```python
test.head()
```

```python
data[['Age','Fare']].median()
```

```python
test[['Age','Fare']].median()
```

```python
#replacing null values with mode
data['Age'].fillna("28", inplace=True)
data['Fare'].fillna("14.4542", inplace=True)
test['Age'].fillna('27',inplace=True)
test['Fare'].fillna('14.4542',inplace=True)
```

```python
data['Embarked'].mode()
```

```python
data['Embarked'].fillna("S", inplace=True)
```

```python
data.info()
```

```python
#ONE HOT ENCODING WE WILL PERFORM ON SEX  AND EMBARKED FOR TRAIN AND TEST DATA BOTH
data['Sex']=data['Sex'].apply({'male':0,'female':1}.get) 
data['Embarked']=data['Embarked'].apply({'S':0,'Q':1, 'C':2}.get) 
test['Sex']=test['Sex'].apply({'male':0,'female':1}.get) 
test['Embarked']=test['Embarked'].apply({'S':0,'Q':1, 'C':2}.get)
```

```python
#DIVIDING THE INDEPENDENT DATA AND DEPENDENT DATA FOR TESTING AND TRAINING
x=data.drop(['Survived'],axis=1)
y=data['Survived']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
from sklearn.ensemble import RandomForestClassifier
forest=RandomForestClassifier()
```

```python
forest.fit(x_train,y_train)
```

```python
forest_pred=forest.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
cm=confusion_matrix(y_test,forest_pred)
ac=accuracy_score(y_test,forest_pred)
```

```python
print(cm)
```

```python
print(ac)
```

```python
test_pred=forest.predict(test)
```

```python
test_id=test['PassengerId']
```

```python
df=pd.DataFrame({'PassengerId':test_id.values,
              "Survived":test_pred
})
```
